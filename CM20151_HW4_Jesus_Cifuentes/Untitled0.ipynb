```python
import matplotlib.pyplot as plt
import numpy as np
import wave
import sys


spf = wave.open('b.wav','r')

#Extract Raw Audio from Wav File
signal = spf.readframes(-1)
signal = np.fromstring(signal, 'Int16')
fs = spf.getframerate()

#If Stereo
#if spf.getnchannels() == 2:
#    print 'Just mono files'
#    sys.exit(0)

t = linspace(0, len(signal)/fs, num=len(signal))
plt.figure(1)
plt.title('Signal Wave...')
plt.plot( t,signal)
```

```python
from scipy.fftpack import fft, fftfreq
signalf = fft(signal)
plt.plot( signalf)
n = len(signal)

freq = fftfreq(n, 6/n) # Recuperamos las frecuencias
plot(freq, np.abs(signalf))
```
